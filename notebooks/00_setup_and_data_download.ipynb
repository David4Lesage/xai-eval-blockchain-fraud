# 00 - Environment setup and dataset download

This notebook verifies that the Python environment is ready and that the
two required datasets are present on disk. Run it first.

### What it does
- Prints package versions to help diagnose environment issues.
- Checks for the Elliptic Bitcoin and Ethereum Fraud Detection datasets.
- Tries to fetch them automatically via the Kaggle CLI when available.
- Prints manual instructions when automatic download is not possible.


```python
from xai_blockchain_framework.config import CONFIG, PATHS, set_seed
from xai_blockchain_framework.data.download import (
    check_datasets_present,
    print_download_instructions,
    try_kaggle_download,
)

set_seed()
print(f'Repository root: {PATHS.repo_root}')
print(f'Random seed:     {CONFIG.random_seed}')
```

## Package versions

```python
import sys
import importlib

print(f'Python: {sys.version.split()[0]}')
packages = [
    ('numpy', 'numpy'),
    ('pandas', 'pandas'),
    ('scikit-learn', 'sklearn'),
    ('lightgbm', 'lightgbm'),
    ('torch', 'torch'),
    ('torch-geometric', 'torch_geometric'),
    ('shap', 'shap'),
    ('lime', 'lime'),
    ('openai', 'openai'),
]
for display_name, import_name in packages:
    try:
        module = importlib.import_module(import_name)
        version = getattr(module, '__version__', 'installed (no __version__ attribute)')
        print(f'  {display_name}: {version}')
    except ImportError:
        print(f'  {display_name}: NOT INSTALLED')
```

## Dataset presence check

```python
status = check_datasets_present()
print(status)

if not all(status.values()):
    print()
    print('At least one dataset is missing; attempting Kaggle CLI download...')
    ok = try_kaggle_download()
    if not ok:
        print()
        print_download_instructions()
else:
    print('Both datasets are ready.')
```
