# 01a - Exploratory Data Analysis: Elliptic Bitcoin

High-level statistics and class distribution for the Elliptic dataset.


```python
import matplotlib.pyplot as plt
import pandas as pd

from xai_blockchain_framework.config import PATHS, set_seed
from xai_blockchain_framework.data import load_elliptic

set_seed()
data = load_elliptic()
print(f'{len(data.features):,} transactions, {data.n_features} features')
```

## Class distribution

```python
labels_str = data.labels.astype(str)
counts = labels_str.value_counts().rename({'1': 'illicit', '2': 'licit'})
print(counts)

ax = counts.plot(kind='bar', color=['#95a5a6', '#2ecc71', '#e74c3c'], figsize=(7, 4))
ax.set_ylabel('Number of transactions')
ax.set_title('Elliptic Bitcoin - class distribution', fontweight='bold')
plt.tight_layout()
plt.savefig(PATHS.figures_dir / 'elliptic_class_distribution.png', dpi=150)
plt.show()
```

## Temporal distribution

```python
ts_counts = data.time_steps.value_counts().sort_index()
ax = ts_counts.plot(kind='line', figsize=(10, 4))
ax.set_xlabel('Time step')
ax.set_ylabel('Transactions')
ax.set_title('Elliptic Bitcoin - transactions per time step', fontweight='bold')
plt.tight_layout()
plt.savefig(PATHS.figures_dir / 'elliptic_temporal.png', dpi=150)
plt.show()
```
