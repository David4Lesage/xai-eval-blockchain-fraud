# 01b - Exploratory Data Analysis: Ethereum Fraud Detection

Summary statistics and class distribution for the Ethereum dataset.


```python
import matplotlib.pyplot as plt
import seaborn as sns

from xai_blockchain_framework.config import PATHS, set_seed
from xai_blockchain_framework.data import load_ethereum

set_seed()
data = load_ethereum()
print(f'{len(data.features):,} addresses, {data.n_features} features')
```

## Class distribution

```python
counts = data.labels.map({0: 'legit', 1: 'fraud'}).value_counts()
print(counts)

ax = counts.plot(kind='bar', color=['#2ecc71', '#e74c3c'], figsize=(6, 4))
ax.set_ylabel('Number of addresses')
ax.set_title('Ethereum Fraud Detection - class distribution', fontweight='bold')
plt.tight_layout()
plt.savefig(PATHS.figures_dir / 'ethereum_class_distribution.png', dpi=150)
plt.show()
```

## Feature correlation matrix

```python
corr = data.features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap='RdBu_r', center=0, square=True, cbar_kws={'shrink': 0.5})
plt.title('Ethereum - feature correlations', fontweight='bold')
plt.tight_layout()
plt.savefig(PATHS.figures_dir / 'ethereum_correlations.png', dpi=150)
plt.show()
```
